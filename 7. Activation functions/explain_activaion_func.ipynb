# Model before applying the Activation Function

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
class before_activation(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear_layer_stack = nn.Sequential(
            nn.Linear(in_features=2, out_features=128),
            nn.Linear(in_features=128, out_features=128),
            nn.Linear(in_features=128, out_features=1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.linear_layer_stack(x)
model = before_activation().to(device)
```

```python
model
```

```python
total_params = sum(
	param.numel() for param in model.parameters()
)
total_params
```

```python
# loss and optimizer
loss_fc = nn.BCELoss()

optimizer = torch.optim.Adam(params = model.parameters(),
                            lr = 0.01)
```

```python
# turning data into tensors
X_ten = torch.from_numpy(X).type(torch.float)
y_ten = torch.from_numpy(y).type(torch.float)
```

```python
epochs = 500

train_loss_values = []
test_loss_values = []
epoch_count = []

for epoch in range(epochs):

  model.train()

  # forward pass
  y_logits = model(X_ten).squeeze()
  y_pred = torch.round(torch.sigmoid(y_logits))

  # calculate the loss/accuarcy
  loss = loss_fc(y_logits, y_ten)

  # optimizer zero grad
  optimizer.zero_grad()

  # backpropagation
  loss.backward()

  # optimizer step (perform gradient descent)
  optimizer.step()

  # testing
  model.eval()
  with torch.inference_mode():

    # forward pass
    test_logits = model(X_ten).squeeze()
    test_pred = torch.round(torch.sigmoid(test_logits))

    # calculate the loss/accuarcy
    test_loss = loss_fc(test_logits, y_ten)


# lets print out what is happening
    if epoch % 10 == 0:
        epoch_count.append(epoch)
        train_loss_values.append(loss.detach().numpy())
        test_loss_values.append(test_loss.detach().numpy())
        print(f"Epoch: {epoch} | Loss: {loss} | Validation Loss = {test_loss.item()}")
```

```python
# 1. Generate a grid of points
x1_range = np.linspace(-1, 0.6, 200)
x2_range = np.linspace(-1, 0.6, 200)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
grid = np.c_[x1_grid.ravel(), x2_grid.ravel()]
grid_tensor = torch.tensor(grid, dtype=torch.float32)

# 2. Predict probabilities for the grid points
with torch.no_grad():
    grid_preds = model(grid_tensor).numpy()

# 3. Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(x1_grid, x2_grid, grid_preds.reshape(x1_grid.shape), levels=[0, 0.5, 1], alpha=0.8, cmap='coolwarm')
plt.colorbar(label='Prediction Probability')

# 4. Overlay the data points
plt.scatter(X[:, 0], X[:, 1], c=y.ravel(), edgecolor='k', cmap='coolwarm', s=50, label='Data Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary')
plt.legend()
plt.grid()
plt.show()
```

# After applying the Activation Function

```python
class after_activation(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear_layer_stack = nn.Sequential(
            nn.Linear(in_features=2, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.linear_layer_stack(x)
model1 = after_activation().to(device)
```

```python
model1
```

```python
# loss and optimizer
loss_fc = nn.BCELoss()

optimizer = torch.optim.Adam(params = model1.parameters(),
                            lr = 0.01)
```

```python
epochs = 500

train_loss_values = []
test_loss_values = []
epoch_count = []

for epoch in range(epochs):

  model1.train()

  # forward pass
  y_logits = model1(X_ten).squeeze()
  y_pred = torch.round(torch.sigmoid(y_logits))

  # calculate the loss/accuarcy
  loss = loss_fc(y_logits, y_ten)

  # optimizer zero grad
  optimizer.zero_grad()

  # backpropagation
  loss.backward()

  # optimizer step (perform gradient descent)
  optimizer.step()

  # testing
  model1.eval()
  with torch.inference_mode():

    # forward pass
    test_logits = model1(X_ten).squeeze()
    test_pred = torch.round(torch.sigmoid(test_logits))

    # calculate the loss/accuarcy
    test_loss = loss_fc(test_logits, y_ten)


# lets print out what is happening
    if epoch % 10 == 0:
        epoch_count.append(epoch)
        train_loss_values.append(loss.detach().numpy())
        test_loss_values.append(test_loss.detach().numpy())
        print(f"Epoch: {epoch} | Loss: {loss} | Validation Loss = {test_loss.item()}")
```
