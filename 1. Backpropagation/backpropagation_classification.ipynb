```python
def initialize_parameters(layer_dim):
    np.random.seed(42)
    parameters = {}
    L = len(layer_dim)

    for i in range(1, L):
        parameters["W" + str(i)] = np.ones((layer_dim[i-1], layer_dim[i]))*0.1
        parameters["b" + str(i)] = np.zeros((layer_dim[i], 1))

    return parameters
```

```python
# sigmoid function
def sigmoid(z):
    A = 1 / (1+ np.exp(-z))
    return A
```

```python
def linear_forward(A_pre, W, b):

    z = np.dot(W.T, A_pre) + b

    A = sigmoid(z) 

    return A
```

```python
def L_layer_forward(X, parameters): 
    A = X
    L = len(parameters) // 2 ## number of layers in the neural network

    for i in range(1, L+1):
        A_pre = A   # A_pre is the input
        W1 = parameters["W" + str(i)]
        b1 = parameters["b" + str(i)]

        print("A" + str(i-1) + ": ", A_pre)
        print("W" + str(i) + ": ", W1)
        print("b" + str(i) + ": ", b1)
        print("--"*20)

        A = linear_forward(A_pre, W1, b1)
        print("A" + str(i) + ": ", A )
        print("**"*20)
    return A, A_pre
```

```python
def update_parameters(y, y_hat, parameters, X, A1):
    parameters["W2"][0][0] = parameters["W2"][0][0] - 0.001*(-1 * (y-y_hat) * A1[0][0])
    parameters["W2"][1][0] = parameters["W2"][1][0] - 0.001*(-1 * (y-y_hat) * A1[1][0])
    parameters["b2"][0][0] = parameters["b2"][0][0] - 0.001*(-1 * (y-y_hat))
    
    parameters["W1"][0][0] = parameters["W1"][0][0] - 0.001*(-1 * (y-y_hat) * parameters["W2"][0][0] * A1[0][0] * (1-A1[0][0]) * X[0][0])
    parameters["W1"][0][1] = parameters["W1"][0][1] - 0.001*(-1 * (y-y_hat) * parameters["W2"][0][0] * A1[0][0] * (1-A1[0][0]) * X[1][0])
    parameters["b1"][0][0] = parameters["b1"][0][0] - 0.001*(-1 * (y-y_hat) * parameters["W2"][0][0] * A1[0][0] * (1-A1[0][0]))
    
    parameters["W1"][1][0] = parameters["W1"][1][0] - 0.001*(-1 * (y-y_hat) * parameters["W2"][1][0] * A1[1][0] * (1-A1[1][0]) * X[0][0])
    parameters["W1"][1][1] = parameters["W1"][1][1] - 0.001*(-1 * (y-y_hat) * parameters["W2"][1][0] * A1[1][0] * (1-A1[1][0]) * X[1][0])
    parameters["b1"][1][0] = parameters["b1"][1][0] - 0.001*(-1 * (y-y_hat) * parameters["W2"][1][0] * A1[1][0] * (1-A1[1][0]))
    
update_parameters(y, y_hat, parameters, X, A1)
```

```python
import torch
from torch import nn
torch.__version__
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
# turn data into tensors
X = torch.from_numpy(X).type(torch.float)
y = torch.from_numpy(y).type(torch.float)
```

```python
X, y
```

```python
# setting up the device agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# Now lets construct a model class

class student(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear_layer_stack = nn.Sequential(
        nn.Linear(in_features=2, out_features=2),
        nn.Linear(in_features=2, out_features=1))

  def forward(self, x):
    return self.linear_layer_stack(x)

model = student().to(device)
```

```python
model
```

```python
model.state_dict()
```

```python
# total_trainable_parameters = [2x2,+ 2 bias], [2x1,+ 1 bias] => 4+2+2+1 => 9
total_params = sum(
	param.numel() for param in model.parameters()
)
total_params
```

```python
# Given weights and biases
weights = [
    torch.tensor([[0.1, 0.1], [0.1, 0.1]], dtype=torch.float32),  # fc1 weight
    torch.tensor([0., 0.], dtype=torch.float32),                 # fc1 bias
    torch.tensor([[0.1], [0.1]], dtype=torch.float32),           # fc2 weight
    torch.tensor([0.], dtype=torch.float32)                      # fc2 bias
]
```

```python
weights
```

```python
# Assign the weights to the Sequential layers
with torch.no_grad():
    model.linear_layer_stack[0].weight.copy_(weights[0])  # First Linear Layer (fc1) weights
    model.linear_layer_stack[0].bias.copy_(weights[1])    # First Linear Layer (fc1) bias
    model.linear_layer_stack[1].weight.copy_(weights[2].T)  # Second Linear Layer (fc2) weights
    model.linear_layer_stack[1].bias.copy_(weights[3])    # Second Linear Layer (fc2) bias

# Verify the assigned weights
print("Layer 1 weights:", model.linear_layer_stack[0].weight)
print("Layer 1 bias:", model.linear_layer_stack[0].bias)
print("Layer 2 weights:", model.linear_layer_stack[1].weight)
print("Layer 2 bias:", model.linear_layer_stack[1].bias)
```

```python
# loss_fc and optimizer
loss_fc = nn.BCELoss() 

optimizer = torch.optim.Adam(params = model.parameters(),
                            lr = 0.001)
```

```python
# building a training and evaluation loop

epochs = 100

for epoch in range(epochs):

  model.train()

  # forward pass
  y_logits = model(X).squeeze()
  y_pred = torch.round(torch.sigmoid(y_logits))

  # calculate the loss/accuarcy
  loss = loss_fc(y_logits, y)

  # optimizer zero grad
  optimizer.zero_grad()

  # backpropagation
  loss.backward()

  # optimizer step (perform gradient descent)
  optimizer.step()

  # testing
  model.eval()
  with torch.inference_mode():

    # forward pass
    test_logits = model(X).squeeze()
    test_pred = torch.round(torch.sigmoid(test_logits))

    # calculate the loss/accuarcy
    test_loss = loss_fc(test_logits, y)


# lets print out what is happening
    if epoch % 10 == 0:
        print(f"Epoch: {epoch} | Loss: {loss:.5f}")
```
